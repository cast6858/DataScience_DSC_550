```python
# 7 Find the five publishers with the highest total global sales. Note: You will need to calculate the total global sales for each publisher to do this.
#Grouping by publisher targeting global sales
#Taking the sum then trageting the largest
VGS_df.groupby('Publisher')['Global_Sales'].sum().nlargest(5)

```

```python
# 8 Create a new column in the data frame that calculates the percentage of global sales from North America. Display the first five rows of the new data frame.
#Dividing NA sales divided by Global sales
#Turned into a percentage 
#Rounded to the second decmial place
VGS_df['Calculated_NA_Golbal_Percentage'] = (VGS_df['NA_Sales'].astype(float)  / VGS_df['Global_Sales'].astype(float) * 100).round(2)
VGS_df.head(5)
```

```python
# 9 Find the number NaN entries (missing data values) in each column.
#taking sums of nulls in DF
VGS_df.isnull().sum()
```

```python
# 10 Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user score column with the median value.

#Replace all strings with NAN values 
#if value contains string replace with nan
#side note - AZ for caps az for lower case words
#value inside user_score replaced 
#replace all string values regradless of verbage, and replace with NAN
VGS_df['User_Score'] = VGS_df['User_Score'].replace(r'([A-Za-z])' , np.NaN , regex = True)

#Overwrite the user NAN score with median score
VGS_df['User_Score'] = VGS_df['User_Score'].replace( np.NaN , VGS_df['User_Score'].median()) 
VGS_df['User_Score'].head(10)
```
